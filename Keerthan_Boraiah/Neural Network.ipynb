# Data Prep

```python
import pandas as pd
import numpy as np
import os
import gc

with open (r'E:\Documents\University Assignments\Game AI\ce811-the-resistance-main\ce811-the-resistance-main\logs\LoggerBot.log') as f:
    file = f.readlines()
    
L = []
for i in file:
    split = i.split(',')[4:]
    L.append(split)
print(len(L))

df = pd.DataFrame([L[0]])
for i in range(len(L)):
    if i == 0:
        continue
    else:
        df = df.append([L[i]])
        
df.reset_index(inplace = True,drop = True)
df.rename({0:'missions_been_on',1:'failed_missions_been_on',2:'missions_voted_up1',3:'missions_voted_up2',4:'missions_voted_up3',
          5:'missions_voted_up4',6:'missions_voted_up5',7:'missions_voted_up6',8:'missions_voted_down1',9:'missions_voted_down2',
          10:'missions_voted_down3',11:'missions_voted_down4',12:'missions_voted_down5',13:'missions_voted_down6',14:'Spy'},axis = 1,inplace = True)

def changer(x):
    if x == "0\n":
        return 0
    else:
        return 1
df['Spy'] = df['Spy'].apply(changer)

gc.collect()
```

# Neural Network Model

```python
num_inputs=X_train.shape[1]
num_outputs=2
```

```python
import numpy as np 
x_train=df.values[:,:14].astype(np.float32)
y_train=df.values[:,14].astype(np.int32)
print(y_train.shape)
print(y_train[0:6])
print(x_train[0:6])
num_inputs=x_train.shape[1]
num_outputs=2
```

```python
dataset_size=len(x_train)
train_set_size=int(dataset_size*0.7)
x_val,y_val=x_train[train_set_size:],y_train[train_set_size:]
x_train,y_train=x_train[:train_set_size],y_train[:train_set_size]
```

```python
x_train.shape,y_train.shape
```

```python
x_val.shape,y_val.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential(name="my_neural_network")
layer1=layers.Dense(10, activation="tanh", input_shape=(num_inputs,))
model.add(layer1)
layer2=layers.Dense(10, activation="tanh")
model.add(layer2)
layer3=layers.Dense(num_outputs)
model.add(layer3)

print(model(x_train[0:3])
```

```python
model.summary()
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(0.001),
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics='accuracy'
)
```

```python
history = model.fit(
    x_train,
    y_train,
    batch_size=50,
    epochs=120,
    validation_data=(x_val, y_val),verbose=1
)
```

```python
keras.models.save_model(model, "Neural_Network.h5",save_format='h5')
```
