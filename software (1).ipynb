## Extracting the data from zip and concatenating them

## Splitting data

```python
feature_cols = ['Recording timestamp', 'Validity left','Validity right' ,'Sensor', 'Recording duration', 'Fixation point X (MCSnorm)', 'Fixation point Y (MCSnorm)', 'Gaze point right X (MCSnorm)', 'Gaze point right Y (MCSnorm)','Event','Eyetracker timestamp']
X = concatenated_df[feature_cols]
y = concatenated_df['Eye movement type']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train.shape, X_test.shape
```

## Cross validation

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score
from sklearn import linear_model, tree, ensemble
```

```python
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cnt = 1

for train_index, test_index in kf.split(X, y):
    print(f'Fold:{cnt}, Train set: {len(train_index)}, Test set:{len(test_index)}')
    cnt+=1
```

```python
score = cross_val_score(linear_model.LogisticRegression(random_state= 42), X, y, cv= kf, scoring="accuracy")
print(f'Each fold scores: {score}')
print(f'The average score: {"{:.2f}".format(score.mean())}')
```

## Random forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf_classifier = RandomForestClassifier(n_estimators=100)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f}%'.format(accuracy*100))
```

## Hyper-parameter tuning

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5)
```

```python
grid_search.fit(X, y)
```

```python
print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```
